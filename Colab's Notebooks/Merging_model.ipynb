```python
import torch
from peft import AutoPeftModelForCausalLM

adapter = 'Hieu-Pham/Llama2-7B-QLoRA-cooking-text-gen-prompting'
device_map = {"": 0}

trained_model = AutoPeftModelForCausalLM.from_pretrained(
    adapter,
    low_cpu_mem_usage=True,
    return_dict=True,
    torch_dtype=torch.float16,
    device_map=device_map,
)
```

```python
# Merge LoRA with the base model and save the merged model
lora_merged_model = trained_model.merge_and_unload()
lora_merged_model.save_pretrained("merged",safe_serialization=True)

#push merged model to the hub
#lora_merged_model.push_to_hub("merged_model_testing")
```

```python
lora_merged_model.push_to_hub("Llama2-7B-QLoRA-cooking-text-gen-prompting-merged")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('meta-llama/Llama-2-7b-hf')
tokenizer.push_to_hub('Llama2-7B-QLoRA-cooking-text-gen-prompting-merged')
```
